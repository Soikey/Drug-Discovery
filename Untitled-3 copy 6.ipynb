```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.loader import DataLoader
from torch_geometric.nn import GINEConv, BatchNorm, global_mean_pool
from transformers import AutoTokenizer, AutoModel
from rdkit import Chem
from rdkit.Chem import rdmolops
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, classification_report
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python

# ========== Load dataset ==========
file_path = 'bace.csv'
df = pd.read_csv(file_path)
df['class'] = df['class'].apply(lambda x: 1 if x == 'active' else 0)
```

```python
# ========== Initialize ChemBERT ==========
tokenizer = AutoTokenizer.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
chembert_model = AutoModel.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
```

```python

# ========== Tokenize SMILES ==========
def tokenize_smiles(smi):
    tokens = tokenizer(smi, truncation=True, padding='max_length', max_length=128, return_tensors="pt")
    return {
        'input_ids': tokens['input_ids'].squeeze(0),
        'attention_mask': tokens['attention_mask'].squeeze(0)
    }

print("Tokenizing SMILES...")
df['tokens'] = df['canonical_smiles'].apply(tokenize_smiles)
# test_df['tokens'] = test_df['canonical_smiles'].apply(tokenize_smiles)
```

```python
# ========== Atom & bond features ==========
def get_atom_features(mol):
    feats = []
    for atom in mol.GetAtoms():
        feats.append([
            atom.GetAtomicNum(),
            atom.GetDegree(),
            atom.GetFormalCharge(),
            atom.GetHybridization(),
            int(atom.GetIsAromatic())
        ])
    return torch.tensor(feats, dtype=torch.float)

def get_edge_features(mol):
    bonds = mol.GetBonds()
    edge_feats = []
    edge_indices = []
    for bond in bonds:
        i = bond.GetBeginAtomIdx()
        j = bond.GetEndAtomIdx()
        edge_indices.append((i, j))
        edge_indices.append((j, i))  # undirected

        feat = [
            bond.GetBondTypeAsDouble(),
            int(bond.GetIsConjugated()),
            int(bond.IsInRing())
        ]
        edge_feats.append(feat)
        edge_feats.append(feat)  # undirected

    edge_index = torch.tensor(edge_indices, dtype=torch.long).T
    edge_attr = torch.tensor(edge_feats, dtype=torch.float)
    return edge_index, edge_attr

def mol_to_data(row):
    smiles = row['canonical_smiles']
    label = torch.tensor(row['class'], dtype=torch.long)
    tokens = row['tokens']

    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None

    x = get_atom_features(mol)
    edge_index, edge_attr = get_edge_features(mol)

    data = Data(x=x, edge_index=edge_index, edge_attr=edge_attr, y=label)
    data.input_ids = tokens['input_ids']
    data.attention_mask = tokens['attention_mask']
    return data
```

```python
# # ========== Convert to dataset ==========
# dataset = [mol_to_data(row) for _, row in tqdm(df.iterrows(), total=len(df))]
# dataset = [d for d in dataset if d is not None]

# # test_dataset = [mol_to_data(row) for _, row in tqdm(test_df.iterrows(), total=len(test_df))]
# # test_dataset = [d for d in test_dataset if d is not None]

# # ========== Split ==========
# train_data, test_data = train_test_split(dataset, test_size=0.2, stratify=df['class'], random_state=42)

# Step 1: Filter out invalid molecules first
valid_rows = [row for _, row in df.iterrows() if Chem.MolFromSmiles(row['canonical_smiles']) is not None]
df_valid = pd.DataFrame(valid_rows)

# Step 2: Stratified split on filtered df
train_df, test_df = train_test_split(df_valid, test_size=0.2, stratify=df_valid['class'], random_state=42)

# Step 3: Then convert to Data objects
train_dataset = [mol_to_data(row) for _, row in train_df.iterrows()]
test_dataset = [mol_to_data(row) for _, row in test_df.iterrows()]

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16)
```

```python
class CustomGNN(nn.Module):
    def __init__(self, gnn_in, gnn_hidden):
        super().__init__()
        self.node_proj = nn.Linear(gnn_in, gnn_hidden)

        nn1 = nn.Sequential(
            nn.Linear(gnn_hidden, gnn_hidden),
            nn.ReLU(),
            nn.Linear(gnn_hidden, gnn_hidden)
        )
        self.conv1 = GINEConv(nn1, edge_dim=3)

        nn2 = nn.Sequential(
            nn.Linear(gnn_hidden, gnn_hidden),
            nn.ReLU(),
            nn.Linear(gnn_hidden, gnn_hidden)
        )
        self.conv2 = GINEConv(nn2, edge_dim=3)

        self.bn1 = BatchNorm(gnn_hidden)
        self.bn2 = BatchNorm(gnn_hidden)
        self.dropout = nn.Dropout(0.2)


    def forward(self, x, edge_index, edge_attr, batch):
        x = self.node_proj(x)  # Project node features to gnn_hidden dim

        h = F.relu(self.conv1(x, edge_index, edge_attr))
        h = self.bn1(h)
        h = self.dropout(h)

        h_res = h
        h = F.relu(self.conv2(h, edge_index, edge_attr))
        h = self.bn2(h)
        h = self.dropout(h)

        h = h + h_res  # residual connection

        out = global_mean_pool(h, batch)
        return out
```

```python
# ========== Hybrid model ==========
class HybridModel(nn.Module):
    def __init__(self, chembert, gnn_in=5, gnn_hidden=64, bert_hidden=768, out_dim=2):
        super().__init__()
        self.gnn = CustomGNN(gnn_in, gnn_hidden)
        self.chembert = chembert

        self.vector_gate = nn.Sequential(
            nn.Linear(gnn_hidden + bert_hidden, gnn_hidden + bert_hidden),
            nn.Sigmoid()
        )
        self.linear1 = nn.Linear(gnn_hidden + bert_hidden, 128)
        self.linear2 = nn.Linear(128, out_dim)

    def forward(self, data):

        # print(f"data.x.shape = {data.x.shape}")
        # print(f"data.edge_attr.shape = {data.edge_attr.shape}")
        
        x = self.gnn(data.x, data.edge_index, data.edge_attr, data.batch)

        input_ids = torch.stack([d.input_ids for d in data.to_data_list()]).to(x.device)
        attention_mask = torch.stack([d.attention_mask for d in data.to_data_list()]).to(x.device)

        chem_out = self.chembert(input_ids=input_ids, attention_mask=attention_mask)
        chembert_out = chem_out.last_hidden_state[:, 0, :]

        combined = torch.cat([x, chembert_out], dim=1)
        gate = self.vector_gate(combined)
        gated = combined * gate

        out = F.relu(self.linear1(gated))
        out = self.linear2(out)

        self.last_gate = gate.detach().cpu()
        return out
```

```python
# ========== Training setup ==========
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = HybridModel(chembert_model).to(device)
# Unfreeze ChemBERTa encoder
# for param in model.chembert.parameters():
#     param.requires_grad = True  # this ensures joint training

optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)

# optimizer = torch.optim.AdamW([
#     {'params': model.chembert.parameters(), 'lr': 1e-5},
#     {'params': model.gnn.parameters(), 'lr': 1e-4},  # or whatever GNN learning rate
#     {'params': model.vector_gate.parameters()},
#     {'params': model.linear1.parameters()},
#     {'params': model.linear2.parameters()},
# ])

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=2, verbose=True)
criterion = nn.CrossEntropyLoss()
```

```python
# ========== Train & eval ==========
def train():
    model.train()
    total_loss, correct = 0, 0
    gates = []
    for data in train_loader:
        data = data.to(device)
        optimizer.zero_grad()
        out = model(data)
        loss = criterion(out, data.y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        pred = out.argmax(dim=1)
        correct += (pred == data.y).sum().item()
        gates.append(model.last_gate)

    acc = correct / len(train_loader.dataset)
    avg_gate = torch.cat(gates).mean(dim=0).numpy()
    return total_loss / len(train_loader), acc, avg_gate

def evaluate(loader):
    model.eval()
    preds, labels = [], []
    with torch.no_grad():
        for data in loader:
            data = data.to(device)
            out = model(data)
            pred = out.argmax(dim=1)
            preds.extend(pred.cpu().numpy())
            labels.extend(data.y.cpu().numpy())

    acc = accuracy_score(labels, preds)
    cm = confusion_matrix(labels, preds)
    roc = roc_auc_score(labels, preds)
    report = classification_report(labels, preds, target_names=['inactive', 'active'])
    return acc, cm, roc, report, preds, labels
```

```python

# ========== Run training ==========
epochs = 20
all_gates = []
losses = []
rocs = []
trains = []
tests = []
best_roc = 0

for epoch in range(epochs):
    loss, train_acc, avg_gate = train()
    test_acc, cm, roc, report, preds, labels = evaluate(test_loader)
    scheduler.step(roc)

    losses.append(loss)
    rocs.append(roc)
    trains.append(train_acc)
    tests.append(test_acc)
    all_gates.append(avg_gate)

    if roc > best_roc:
        best_roc = roc
        torch.save(model.state_dict(), 'best_model3_4.pt')
        print(f"✅ Saved new best model at epoch {epoch+1} with ROC-AUC: {roc:.4f}")

    print(f"ChemBERTa grad check: {model.chembert.encoder.layer[0].intermediate.dense.weight.requires_grad}")

    print(f"Epoch {epoch+1}: Loss={loss:.4f}, Train Acc={train_acc:.4f}, Test Acc={test_acc:.4f}, ROC-AUC={roc:.4f}")
    print(f"Confusion Matrix:\n{cm}")
    print(f"Classification Report:\n{report}")
    print(f"Mean Gate (first 10 dims): {avg_gate[:10]}")



# ========== Plot results ==========
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(losses, label='Train Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(rocs, label='Test ROC-AUC')
plt.xlabel('Epoch')
plt.ylabel('ROC-AUC')
plt.title('Test ROC-AUC')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(trains, label='Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Training Accuracy')
plt.title('Training Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(tests, label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

all_gates = np.array(all_gates)
plt.figure(figsize=(10, 5))
for i in range(min(10, all_gates.shape[1])):
    plt.plot(all_gates[:, i], label=f'Gate dim {i}')
plt.xlabel('Epoch')
plt.ylabel('Mean gate value')
plt.title('Gate value evolution over epochs')
plt.legend()
plt.show()

```
